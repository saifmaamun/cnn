```python
xgb = XGBClassifier()

xgb.fit(X_train_features, y_train)

xgb_pred = xgb.predict(X_test_features)
accuracy_score(y_test, xgb_pred)
```

```python

rf = RandomForestClassifier()

rf.fit(X_train_features, y_train)
rf_pred = rf.predict(X_test_features)
accuracy_score(y_test, rf_pred)
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train_features, y_train)
knn_pred = knn.predict(X_test_features)
accuracy_score(y_test, knn_pred)
```

```python
svm = SVC()

svm.fit(X_train_features, y_train)
svm_pred = svm.predict(X_test_features)
accuracy_score(y_test, svm_pred)
```

```python
lr = LogisticRegression(max_iter=500)

lr.fit(X_train_features, y_train)

lr_pred = lr.predict(X_test_features)
accuracy_score(y_test, lr_pred)
```
