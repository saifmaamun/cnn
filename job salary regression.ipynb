```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
# import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("job_salary_prediction_dataset.csv")
df.shape
```

```python
df.head()
```

```python
print("************************************")
print(df["job_title"].value_counts())
print("************************************")
print("************************************")
print(df["education_level"].value_counts())
print("************************************")
print("************************************")
print(df["industry"].value_counts())
print("************************************")
print("************************************")
print(df["company_size"].value_counts())
print("************************************")
print("************************************")
print(df["location"].value_counts())
print("************************************")
print("************************************")
print(df["remote_work"].value_counts())
print("************************************")
```

```python
dummy =pd.get_dummies(
    df[["job_title","education_level", "industry","company_size","location","remote_work"]], 
    drop_first=True, 
    dtype=int
)
dummy.head()
```

```python
ndf = pd.concat([df,dummy], axis='columns')
ndf.head()
```

```python
ndf=ndf.drop(["job_title","education_level", "industry","company_size","location","remote_work"], axis='columns')
ndf.head()
```

```python
X= ndf.drop("salary", axis = 'columns')
X.head()
```

```python
y = ndf.salary
y.head()
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted=model.predict(X_test)
y_predicted
```

```python
r2 = r2_score(y_test, y_predicted)
r2
```

```python
mse = mean_squared_error(y_test, y_predicted)
mse
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True)
X_test_poly = poly.fit_transform(X_test)
X_train_poly = poly.fit_transform(X_train)
```

```python
model = LinearRegression()
model.fit(X_train_poly,y_train)
```

```python
model.score(X_test_poly,y_test)
```

```python
y_predicted=model.predict(X_test_poly)
r2 = r2_score(y_test, y_predicted)
r2
```

```python
mse = mean_squared_error(y_test, y_predicted)
(mse**(.5)/250000)*100
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
```
